```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd

# Messdaten: Pändellänge und Periodendauer

l_mess= np.array([0.2, 0.4, 0.6, 0.8])
T_mess= np.array([0.9, 1.25, 1.55, 1.8])


# Funktion zur Berechnung der Periodendauer T
def pendel_modell(l, g):
    return 2 * np.pi * np.sqrt(l / g)
popt, pcov = curve_fit(pendel_modell, l_mess, T_mess, p0 = [9.81])

#Plotten der Messdaten und des Modells
plt.errorbar(l_mess, T_mess, yerr=0.05 , fmt='ok', label = f'fit (g= {popt[0]:.5f} m/s²)')
l_plot = np.linspace(0.1, 1.1, 100)
plt.plot(l_plot, pendel_modell(l_plot, popt[0]), label = f'Modell (g= {popt[0]:.5f} m/s²)')
plt.xlabel('Länge l [m]')
plt.ylabel('Periodendauer T [s]')
plt.legend()
plt.grid(True)
plt.show()

# nnun können wir die berechnete Erdbeschleunigung g ausgeben
print(f"Erdbeschleunigung g = {popt[0]:.5f} m/s²")
print(f"Fehler der Erdbeschleunigung: {np.sqrt(pcov[0,0]):.5f} m/s²")
```
